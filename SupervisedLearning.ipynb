**Supervised Learning is a machine learning technique where the model learns using data that already contains the correct answers.**

**Components of Supervised Learning**
- Features/independent variable/inputs/ predictors/attributes - inputs used by the machine to make predictions

- targets/label/output/dependent variable/response variable - the value we want to predict

Train the model using prior examples and then it learns from patterns..

**Applications of Supervised learning**
Healthcare - predict sicknesses
Finance
Education
Marketing
Transportation

**Types of supervised learning**
- Regression - predicts numbers
- Classification - Predicts categories
Classification can be defined into various types
- Binary classification - this kind of classification has only 2 possible outcomes
Yes = 1
No =  0

Yes / No

Pass / Fail

True / False

Spam / Not Spam

Survived / Died

- Multiclass classification - 3 or more possible outcomes
Animal classification
